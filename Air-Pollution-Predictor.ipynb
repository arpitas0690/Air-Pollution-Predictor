```python
dfs = []  # List to store individual DataFrames

for file_path in file_paths:
    df = pd.read_csv(file_path)
    dfs.append(df)

#Concatenate the DataFrames into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)
combined_df
```

```python
#Now, I want to match in the Long/Lat data for the counties using data from: Pareto Software, LLC, the owner of Simplemaps.com.
# See license text in Resources folder for more details regarding citation.
data_file = "./Resources/aqi_daily_1980_to_2021.csv"

data_file_df = pd.read_csv(data_file)
data_file_df.head()
```

```python

geo_data_file = data_file_df.rename(columns={"State Name":"State", "County Name":"County"})

geo_data_file = geo_data_file.drop_duplicates(subset=['State', 'County'])

geo_data_file.head()
```

```python
merge_df = pd.merge(combined_df, geo_data_file, on=["State","County"], how='left')
merge_df
```

```python
#There are several thousand rows with missing data, identifying what they are

# Assuming 'Longitude' is the column containing longitude data
rows_with_missing_longitude = merge_df[merge_df['Latitude'].isnull()]

# Display the rows with missing longitude
selected_columns = ["County","State","Year"]
print(rows_with_missing_longitude[selected_columns])
```
